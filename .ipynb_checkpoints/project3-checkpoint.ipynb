```python
import yfinance as yf
import matplotlib.pyplot as plt
# amzn for Amazon, aapl for Apple, msft for Microsoft, etc
ticker = input('Enter stock ticker: ') 
#Valid periods: 1d,5d,1mo,3mo,6mo,1y,2y,5y,10y,ytd,max Either Use period parameter or use start and end.
period = input('Enter the period you want the data downloaded: ') 

data = yf.download(ticker,period=period)
data.head()

```

```python
plt.plot(data['Adj Close']) # We use adjusted closing price
plt.show()
close_px = data['Adj Close']

rets = close_px.pct_change() # The percent change in stock price
rets.plot()
```
